```python
df = pd.read_excel('./boston housing.xls')
df
```

```python
x = df.drop(columns=['MEDV', 'CAT. MEDV'])
y = df['MEDV']
```

```python
# 모델 생성

X_train, X_test, y_train, y_test = train_test_split(x, y)

params = {'n_estimators': [100], 'max_depth': [5,6,7, 8, 10, 12], 'min_samples_leaf': [7,8,9, 12, 18],
          'min_samples_split': [8, 15,16,17, 20]}

clf = RandomForestRegressor(n_estimators=100)
grid_clf = GridSearchCV(clf, param_grid=params, cv=2, n_jobs=-1)  # -1 은 cpu를 다 쓴다는 의미
grid_clf.fit(X_train, y_train)

print(f"최적의 파라미터\n{grid_clf.best_params_}")
print(f"최고 예측 정확도: {grid_clf.best_score_}")
```
