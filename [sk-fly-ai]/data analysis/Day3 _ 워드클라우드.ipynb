```python
user_agent = {"User-Agent": "Mozilla/5.0 (Macintosh;Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
from bs4 import BeautifulSoup
import requests
import time
questions = []
```

```python
# 프로그래머스 질의응답 사이트 크롤링
for i in range(1, 30):
  #페이지네이션 적용
  res = requests.get("https://qna.programmers.co.kr/?page={}".format(i), user_agent)
  soup = BeautifulSoup(res.text, "html.parser")
  ul_tags = soup.find_all("ul", "question-tags")
  for ul in ul_tags:
    li_tags = ul.find_all("li")
    for li in li_tags:
      #빈칸이 너무많아 strip()으로 처리
      questions.append(li.text.strip())
  time.sleep(0.5)
```

```python
# 형태소 분석 라이브러리 다운로드
!pip install konlpy
```

```python
# 형태소분석
# 횟수를 기반으로 딕셔너리 생성
from collections import Counter
# 문장에서 명사를 추출하는 형태소 분석 라이브러리
from konlpy.tag import Hannanum
words = []
hannanum = Hannanum()
#문장이 아니라 지금은 리스트이기 때문에 for문을 돌면서 각각의 형태소를 분석.
for question in questions:
  nouns = hannanum.nouns(question) #한 번 반복할 때 나오는 명사들
  words += nouns #word에 누적.

# counter를 이용해 각 단어의 개수.
counter = Counter(words)
counter
```

```python
# 리팩토링
from collections import Counter
from konlpy.tag import Hannanum
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def update_stopwords(additional_stopwords):
    from wordcloud import STOPWORDS
    STOPWORDS.update(additional_stopwords)
    return STOPWORDS

def extract_nouns(text, stopwords):
    hannanum = Hannanum()
    nouns = hannanum.nouns(text)
    return [word for word in nouns if word not in stopwords]

def create_wordcloud(counter, font_path, mask=None, background_color='white', width=1000, height=1000):
    wordcloud = WordCloud(
        font_path=font_path,
        mask=mask,
        background_color=background_color,
        width=width,
        height=height
    )
    img = wordcloud.generate_from_frequencies(counter)
    plt.axis('off')
    plt.imshow(img)
    plt.show()

# 불용어 설정
stopwords = update_stopwords(["더 중앙", "중앙 플러스", "더", "중앙일보", "의", "를", "는", "가", "에", "이", "을", "중앙", "플러스", "으로", "에서", "은", "등", "로"])

# 명사 추출
nouns = extract_nouns(article_text, stopwords)

# 빈도수 계산
counter = Counter(nouns)

# 기본 워드클라우드 생성 및 표시
create_wordcloud(counter, '/usr/share/fonts/truetype/nanum/NanumBarunGothicBold.ttf')

# 마스크 처리 후 워드클라우드 생성 및 표시
word_mask = np.array(Image.open('fake.png'))
create_wordcloud(counter, '/usr/share/fonts/truetype/nanum/NanumBarunGothicBold.ttf', mask=word_mask, background_color='ivory')
```
