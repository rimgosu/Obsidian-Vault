```python
q1 = df['age'].quantile(.25)
q3 = df['age'].quantile(.75)
q1, q3
```

```python
iqr = q3 - q1
iqr
```

```python
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
upper, lower
```

```python
cond1 = df['age'] <= upper
cond2 = df['age'] >= lower
```

```python
df[cond1 & cond2].shape
```

```python
mean_age = df['age'].mean()
std_age = df['age'].std()
mean_age, std_age
```

```python
znorm = (df['age'] - mean_age) / std_age
znorm
```

```python
cond_z = znorm > 3
df[cond_z].shape
```
